```python
from transformers import pipeline, AutoModelForQuestionAnswering, BertTokenizer
import transformers
```

```python
finbert=AutoModelForQuestionAnswering.from_pretrained('deepset/bert-base-cased-squad2')
tokenizer=BertTokenizer.from_pretrained('deepset/bert-base-cased-squad2')
```

```python
nlp=pipeline('question-answering',model=finbert,tokenizer=tokenizer)
context = """
ABC Corp is a publicly traded technology company that specializes in cloud computing services. The company reported its financial results for the fiscal year 2023, highlighting significant growth in both revenue and net income. The following key financial metrics were disclosed:

- Total Revenue: $12 billion, a 20% increase from the previous year.
- Net Income: $2.5 billion, a 25% increase from the previous year.
- Earnings Per Share (EPS): $5.00, up from $4.00 in the previous year.
- Operating Expenses: $8 billion, which includes $1 billion in research and development (R&D) costs.
- Free Cash Flow: $3 billion, compared to $2.5 billion last year.
- Total Assets: $30 billion, with total liabilities of $10 billion.
- Stock Price: $150 per share, compared to $120 per share a year ago.
- Number of Employees: 15,000.
"""
questions = [
    "What was ABC Corp's total revenue for the fiscal year 2023?",
    "How much did ABC Corp's net income increase in percentage compared to the previous year?",
    "What was the Earnings Per Share (EPS) for ABC Corp in the fiscal year 2023?",
    "By how much did the EPS increase from the previous year?",
    "What were ABC Corp's operating expenses in the fiscal year 2023?",
    "How much did ABC Corp spend on research and development (R&D) in 2023?",
    "What was the free cash flow for ABC Corp in 2023, and how does it compare to the previous year?",
    "What is the total value of ABC Corp's assets as of the end of the fiscal year 2023?",
    "What are the total liabilities of ABC Corp as of the end of the fiscal year 2023?",
    "What was ABC Corp's stock price at the end of the fiscal year 2023?",
    "How much did the stock price increase compared to the previous year?",
    "How many employees did ABC Corp have at the end of the fiscal year 2023?",
    "What factors might contribute to ABC Corp’s 20% increase in total revenue?",
    "Given the net income and total revenue figures, what is ABC Corp’s net profit margin for the fiscal year 2023?",
    "Considering the free cash flow of $3 billion and net income of $2.5 billion, what can be inferred about ABC Corp’s cash conversion efficiency?",
    "Based on the EPS and stock price, what is the Price-to-Earnings (P/E) ratio for ABC Corp at the end of the fiscal year 2023?",
    "How does this P/E ratio compare to the industry average of 30?"
]
#tokenizer.encode(questions, truncation=True, padding=True)
```

```python
results=nlp(question=questions,context=context)
for i in range(len(results)):
  print(results[i])
  print('\n')
```

```python
for i in range(len(results)):
  print('Question : ',questions[i])
  print('Answer   : ',results[i]['answer'])
```
